```python
plt.figure(figsize=(12,8))
plt.imshow(tumor)
```

```python
tumor = tumor/255.0
tumor
```

```python
tumor.shape
```

```python
X = tumor.reshape(-1,3)
X.shape
```

# Segmentation - K Means

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans (n_clusters = 3)
km.fit(X)
```

```python
img_seg = km.cluster_centers_
img_seg = img_seg[km.labels_]
img_seg
```

```python
img_seg = img_seg.reshape(tumor.shape)
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(img_seg)
```

# Segmentation - MBK



```python
from sklearn.cluster import MiniBatchKMeans
```

```python
kmeans = MiniBatchKMeans(n_clusters=3, random_state=0, batch_size=6, max_iter=10)
kmeans.fit(X)
```

```python
img_seg = kmeans.cluster_centers_
img_seg = img_seg[kmeans.labels_]
img_seg
```

```python
img_seg = img_seg.reshape(tumor.shape)
```

```python
plt.figure(figsize=(12, 8))
plt.imshow(img_seg)
```
